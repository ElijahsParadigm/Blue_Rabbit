```
# This is a import csv
import os
import pandas as pd
import requests
from pathlib import Path
import json
```

```
taxi_df = pd.read_csv("data/Transformed/taxi/taxi.csv")
taxi_df.head()
```

```
lookup_df = pd.read_csv("data/Transformed/reference/ZoneCoordinatesLookup.csv")
lookup_df.head()
```

```
merged_df = pd.merge(taxi_df, lookup_df[['LocationID', 'latitude', 'longitude']], left_on='PULocationID', right_on='LocationID',how='left')
taxi_df['lat_pickup'] = merged_df['latitude']
taxi_df['long_pickup'] = merged_df['longitude']
taxi_df.to_csv("merged.csv")
taxi_df.head()
```

```
taxi_df.head()
```

```
merged_df = pd.merge(taxi_df, lookup_df[['LocationID', 'latitude', 'longitude']],left_on='DOLocationID', right_on='LocationID', how='left')
taxi_df['lat_dropoff'] = merged_df['latitude']
taxi_df['long_dropoff'] = merged_df['longitude']
taxi_df.to_csv("merged.csv")
```

```
# create pikcup time int 
# create dropoff  time int 
taxi_df['pickup_datetime'] = pd.to_datetime(taxi_df['pickup_datetime'])
taxi_df['dropoff_datetime'] = pd.to_datetime(taxi_df['dropoff_datetime'])
taxi_df["pickup_time_int"] = taxi_df['pickup_datetime'].apply(lambda x: x.strftime("%Y%m%d%H"))
taxi_df["dropoff_time_int"] = taxi_df['dropoff_datetime'].apply(lambda x: x.strftime("%Y%m%d%H"))

```

```
taxi_df.head()
taxi_df.to_csv("merged_cleaned.csv")
```

```
weather_df = pd.DataFrame()
for index, row in lookup_df.iterrows():
    latitude = row['latitude']
    longitude = row['longitude']
    url = f"https://archive-api.open-meteo.com/v1/archive?latitude={latitude}&longitude={longitude}&start_date=2023-01-01&end_date=2023-04-01&hourly=temperature_2m,weathercode&temperature_unit=fahrenheit&windspeed_unit=mph&precipitation_unit=inch&timezone=America%2FNew_York"
    # get all weather data for New York Metro
    response = requests.get(url)
    json_data = response.json()
    try:
        df_temp = pd.DataFrame(json_data['hourly'])
        df_temp['latitude'] = json_data['latitude']
        df_temp['longitude'] = json_data['longitude']
        df_temp['time'] = pd.to_datetime(df_temp['time'])
        df_temp["time_int"] = df_temp['time'].apply(lambda x: x.strftime("%Y%m%d%H"))
        weather_df = weather_df.append(df_temp, ignore_index=True)
    except KeyError:
        continue

print(weather_df.shape)
weather_df.head()


```

```
weather_df.to_csv("weather_cleaned.csv")
```
